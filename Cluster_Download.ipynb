```python
start_dc = npzfile['arr_0']
end_dc = npzfile['arr_1']
start_c = npzfile['arr_2']
end_c = npzfile['arr_3']
```

```python
for k in range(1032,len(start_dc)):
    
    time = pandas.Timestamp.time(start_dc[k])
    date = pandas.Timestamp.date(start_dc[k])
    combine = str(pandas.Timestamp.combine(date,time))[0:19]
    strp = (pandas.Timestamp.strptime(combine, '%Y-%m-%d %H:%M:%S'))
    atime = pandas.Timestamp.time(strp)
    adate = pandas.Timestamp.date(strp)
    start_date = str(adate)+"T"+str(atime)+"Z"
    
    time1 = pandas.Timestamp.time(end_dc[k])
    date1 = pandas.Timestamp.date(end_dc[k])
    combine1 = str(pandas.Timestamp.combine(date1,time1))[0:19]
    strp1 = (pandas.Timestamp.strptime(combine1, '%Y-%m-%d %H:%M:%S'))
    atime1 = pandas.Timestamp.time(strp1)
    adate1 = pandas.Timestamp.date(strp1)
    end_date = str(adate1)+"T"+str(atime1)+"Z"
    

    def download(url, params, file_name):
        # open in binary mode
        with open(file_name, "wb") as file:
            # get request
            response = get(url, params=params)
            # write to file
            file.write(response.content)

#DATASET_ID Options:
#C4_CP_CIS-CODIF_LS_H1_MOMENTS
#C2_CP_PEA_PITCH_SPIN_DEFlux
#C2_CP_PEA_PITCH_SPIN_PSD
#C2_CP_FGM_SPIN
#C2_CP_PEA_MOMENTS

    myurl = 'https://csa.esac.esa.int/csa/aio/product-action'
    query_specs = {'DATASET_ID': 'C2_CP_PEA_PITCH_SPIN_PSD',
                   'START_DATE': start_date,
                   'END_DATE': end_date,
                   'DELIVERY_FORMAT': 'CDF',
                   'NON_BROWSER': '1',
                   'DELIVERY_INTERVAL': 'all',
                   'CSACOOKIE': '260A526A39095D6837215D6522090E337E51057D281B522F255957603F091D6D2A064F6C2A114673650A526A39095D682A4602350B1D506D6509502F3E03'}

    filename = str(date.year)+'_'+str(date.month)+'_dc_'+str(k)

    download(myurl, query_specs, str(filename)+'.tar.gz')

    with tarfile.open(str(filename)+".tar.gz") as tar:
        tarname = tar.getnames()
        tar.extractall()
```
